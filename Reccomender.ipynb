```python
# Reccomender systems = used to recomend items
# IMDB movie site
# 2 datasets. 1. Dataset has the movie id, title 2. movie ratings
import pandas
movies = pandas.read_csv("Movie_Id_Titles.csv")
details = pandas.read_csv("file.csv")

# Merge the 2
data = pandas.merge(movies, details, on = 'item_id')
data
```

```python
# Get the highest rated movie on average
data.groupby("title")['rating'].mean().sort_values(ascending=False)
```

```python
data.groupby("title")['rating'].count().sort_values(ascending=False)
```

```python
ratings = pandas.DataFrame(data.groupby("title")['rating'].mean().sort_values(ascending=False))
# We now get how many users rated the movies
ratings['no_of_ratings'] = data.groupby("title")['rating'].count()
ratings
```

```python
# We now use the pivot table
pivot = data.pivot_table(index = 'user_id', columns='title', values='rating')
pivot.head(50)
# The pivot helps us get some relationship btn the user_id, title, rating
```

```python
# Use the pivot to get recommendations of specific movies
selected = pivot['39 Steps, The (1935)']
# Correlation is relation btn 2 var
# Positive corrrelation = 1 goes up,one goes up
# Negative correlation = 1 goes up, the other goes down

similar = pivot.corrwith(selected)
similar_movies = pandas.DataFrame(similar, columns=['Correlations'])
similar_movies

# Lets see how many ratings the movies have
# We join above DataFrame with ratings['no_of_ratings']
similar_movies = similar_movies.join(ratings['no_of_ratings'])
similar_movies

# Lets only show the ones with over 200 ratings
similar_movies[similar_movies['no_of_ratings'] > 200].sort_values('Correlations', ascending=False).head(20)
```
