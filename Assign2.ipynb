```python
import pandas
data = pandas.read_csv('http://modcom.co.ke/data/datasets/AirlinesCluster.csv')
data.head(5)
```

```python
# we use only FlightMiles	FlightTrans	DaysSinceEnroll
# Come up with 10 clusters
# Create your X
data.dropna(inplace=True)
  
```

```python
array = data.values
X = array[:, 4:7]
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=10, random_state=42)
model.fit(X)
```

```python
means = model.cluster_centers_
clusters = pandas.DataFrame(means, columns= ['FlightMiles', 'FlightTrans', 'DaysSinceEnroll'])
clusters
```

```python
data['no'] = model.labels_
cluster = data[data ['no'] == 6]
cluster 
```
