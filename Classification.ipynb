```python
# http://modcom.co.ke/datasets/data.csv
import pandas
data = pandas.read_csv('data.csv')
data
# The data is about types of cakes and the ingredients
# Flour - Grams, Milk - ml, Sugar - Grams, Butter - Grams, Eggs - Count, Baking Powder - Spoons
# Vanilla - spoons, Salt - range from 0 - 10
# Type is the type of cake that was baked
# We train a computer on below data,to understand what makes......
# a Muffin, Cupcake, Scones
# How do we do that?? Machine Learning
# Classification = Supervised Learning, you have to taste your model.
```

```python
# Step 1 : Split data )columns into 2 portions
array = data.values
X = array[:, 0:8]  # 8 is not counted here
Y = array[:, 8]  # This is only the type column
```

```python
# Step 2: Split to 70% training, 30% testing
from sklearn import *
X_train, X_test, Y_train, Y_test = model_selection.train_test_split(X,Y, test_size=0.3, random_state=42)
# X_train and Y_train becomes the 70%
# X_test and Y_test become the 30%
 
```

```python
# Step 3: Bring in a machine learning model
# We will then train the model with the 70% data
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVR

model = RandomForestClassifier()
model.fit(X_train, Y_train)
# The GaussaianNB learns on all the 35 rows
# It understands what level of ingredients make a muffin, a cupcake and a scone
# The model becomes more knowledgable of the three cakes.
# Learinng happens here
```

```python
# Lets test if the model has learned
predictions = model.predict(X_test)
predictions
print('Model predictions: ', predictions)
# The model predicts the cake types
print('These are the answers: ', Y_test)
# The model got all correct
# That is a 90%

```

```python
# Assume you want to cook a cake but you dont know how to go about it.
# Flour	Milk	Sugar	Butter	Egg	Baking Powder	Vanilla	Salt	
recipe = [[19, 13, 17, 24, 5, 100, 70, 0]]
outcome = model.predict(recipe)
print('You will be preparing a: ', outcome)
# The model identifies that less butter and more baking powder - scone
# The model noticed that more butter would trigger a cupcake
```
