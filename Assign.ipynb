```python
# Health - ML is being used in hospital to predict infections
# Ml helps in preventing health issues
import pandas
data = pandas.read_csv('http://modcom.co.ke/data/datasets/pima.csv')
data.head(50)
```

```python
# Create an ML model to help predict if the patient is diabetic or not
array = data.values
X = array[:, 0:8]
Y = array[:, 8]
```

```python
from sklearn import *
X_train, X_test, Y_train, Y_test = model_selection.train_test_split(X,Y, test_size=0.3, random_state=20)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVR

model = DecisionTreeClassifier()
model.fit(X_train, Y_train)
```

```python
predictions = model.predict(X_test)
predictions
print('Model predictions: ', predictions)
# The model predicts whether the patient is diabetic or not
print('These are the answers: ', Y_test)
```
