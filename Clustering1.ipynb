```python
# Clustering - its grouping items into similar groups
# 50,47,53,55 .......Usage 50
# 90,95,93,99,105..........Usage 100
# 160,170,140 ........Usage 150
# Clustering is unsupervised no testing of model
# Lets say the supermarket wants to launch some offers to its customers
# They need to know which specific customers to target
# Clustering its customers is the solution

import pandas
data = pandas.read_csv('http://modcom.co.ke/datasets/Mall_Customers.csv')
data.head(5)

```

```python
# First drop any empties
data.dropna(inplace=True) # This means remove the empties completely
array = data.values
X = array[:, 2:5]
```

```python
# Split the data
array = data.values
X = array[:, 2:5]
# No Y
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=12, random_state=42)
model.fit(X)
```

```python
# Lets see the clusters
means = model.cluster_centers_
clusters = pandas.DataFrame(means, columns= ['Age', 'Income', 'Score'])
clusters
# Below are the three set means for each cluster, for the three columns
```

```python
# Lets see who is in the cluster
# The labels gives you which cluster every record went.
data['no'] = model.labels_
cluster = data[data ['no'] == 5]
cluster 
```
